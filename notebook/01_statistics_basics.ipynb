```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.normal(loc=50, scale=10, size=1000)
df = pd.DataFrame({'value': data})

#histogram
plt.hist(df['value'], bins=30)
plt.show()

#boxplot
sns.boxplot(x=df['value'])
plt.show()

#density plot
sns.kdeplot(df['value'])
plt.show()

#stats
df.describe()
```
